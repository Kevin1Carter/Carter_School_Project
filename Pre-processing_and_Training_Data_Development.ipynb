# Creating Dummy Variables

```python
dfo=df['avg_price_per_room'] 
dummy_room_values= pd.get_dummies(dfo)
dummy_room_values
df=pd.concat([df,dummy_room_values],axis=1)
del(df['avg_price_per_room'])
```

```python
del(df['Booking_ID'])
del(df['type_of_meal_plan'])
del(df['room_type_reserved'])
del(df['market_segment_type'])
del(df['booking_status'])
y=df['no_of_weekend_nights']
no_response_var_data=df.drop(df.loc[:,['no_of_weekend_nights']],axis=1)
scaled=preprocessing.StandardScaler().fit(no_response_var_data)
scaled
X_scaled=scaled.transform(no_response_var_data)
```

```python
no_response_var_data
```

```python
from sklearn.model_selection import train_test_split

y=y.ravel()
X_train, X_test, y_train, y_tes=train_test_split(X_scaled,y)
print(no_response_var_data.columns)
```

```python
from sklearn import linear_model
from sklearn.metrics import explained_variance_score,mean_absolute_error
lm = linear_model.LinearRegression()
model = lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_prediction=model.predict(X_test)
```

```python
explained_variance_score(y_prediction, y_tes)
```

```python
mean_absolute_error(y_prediction, y_tes)
```
